```python
repo = "https://raw.githubusercontent.com/nicoguaro/matplotlib_styles/master"
style.use("style_sheet.mplstyle")

rc('figure', figsize=(8, 4))
rc('savefig', bbox='tight')
plt.rcParams.update({'font.size': 16})

import numpy as np
```

```python
# gradually decreasing the eigen values
with open("invest_cvx_func_first_set.npy","rb") as f:
    all_data = np.load(f)

labels = ["Frobenius norm", "trace", "1/$\epsilon$-flatness", "Pac-Bayes", "negative entropy","low_pass", "fim"]
fig, ax = plt.subplots()
for i in range(7):
    ax.plot([x/100 for x in range(0,101,5)], all_data[i, :] / all_data[i,:].max(),'o-', label=labels[i], Linewidth=3)

ax.legend(loc='lower left')
ax.set_xlabel("Fraction of eigen values of hessian below 1e-3")
ax.set_ylabel("Sharpness measures (normalized)")
# fig.savefig("inv_cvx_func.png")
```

```python
# mean  eigenvalue
with open("invest_cvx_func_unif_10percent.npy","rb") as f:
    all_data = np.load(f)

labels = ["Frobenius norm", "trace", "1/$\epsilon$-flatness", "Pac-Bayes", "negative entropy","low_pass"]
fig, ax = plt.subplots()

for i in range(6):
    ax.plot([x for x in range(1,101,3)], all_data[i, :]/ all_data[i,:].max(),'o-', label=labels[i], Linewidth=3)

ax.legend(loc='lower left')
ax.set_xlabel("mean eigen value")
ax.set_ylabel("Sharpness measures (normalized)")
ax.set_xlim([100,1])
fig.savefig("inv_cvx_func_unif.png")
```
